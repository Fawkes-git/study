```python
df1 = pd.read_csv(os.path.join(os.getcwd(),'data/train_serial.tsv'), delim_whitespace=True)
df2 = pd.read_csv(os.path.join(os.getcwd(),'data/matsumoto/1/train_marker.tsv'), delim_whitespace=True)

```

```python
df1 = df1.set_index('Time')
df2 = df2.set_index('Time')

data = pd.read_table(os.path.join(os.getcwd(),'data/matsumoto/distance.tsv'))
X = data[["Volt"]].values 
y = data["Dist"].values 
regr = LinearRegression()
regr = regr.fit(X, y)
y_fit = regr.predict(X)
print(regr.coef_)
print(regr.intercept_)
min = data["Dist"][0]
max = data["Dist"][1]
```

```python
volt = df1["Volt"]
df1["Dist"] = volt * regr.coef_ + regr.intercept_
df1["Dist"] = df1["Dist"].apply(lambda x : min if x < min else (max if x > max else x))
df1.plot(y="Dist",figsize=(20,10),title='finger movement')
```

```python
#df1 = df1.round()
df2.plot(y="Marker",figsize=(20,10))
```

```python
import math
df1["Theta"] = df1["Dist"].apply(lambda x : math.degrees(math.acos(x/max)))
df1.plot(y="Theta",figsize=(20,10),label="Theta")
```

```python
ax = df1["Theta"].plot(figsize=(20,10),label="Estimate")
p2 = df2.plot(ax = ax)
plt.xlabel("Time(s)")
plt.ylabel("Joint angle()")

plt.title("Angle estimates - 20s")
plt.legend() 
```

```python
df = pd.concat([df1, df2], axis=1, join='outer')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
df = df.interpolate()
df = df.dropna()
```

```python
df_sum = df
```

```python

df_sum = pd.concat([df, df_sum])
```
