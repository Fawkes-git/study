```python
df = pd.read_csv(os.path.join(os.getcwd(),'data/Acceleration_table.csv'), delim_whitespace=False)
```

```python
stats.f_oneway(df.eat, df.wipe, df.type, df.write, df.fold, df.peg)
```

```python
stats.friedmanchisquare(df.eat, df.wipe, df.type, df.write, df.fold, df.peg)
```

```python
fig, ax = plt.subplots()

bp = ax.boxplot([df.eat, df.wipe, df.type, df.write, df.fold, df.peg],labels=activity)
plt.ylabel("Acceleration($m/s^2$)")
```

```python
stats.ttest_ind(df.wipe,df.fold,equal_var = False)
```

```python
df
```

```python
#正規性の検定 p=0.05以上で正規分布
for act in activity:
    print(act,":",stats.shapiro(df[act]))
```

```python
#等分散の検定 p=0.05以上で等分散
stats.bartlett(df.eat, df.wipe, df.type, df.write, df.fold, df.peg)
```

```python
#外れ値の検定
from outliers import smirnov_grubbs as grubbs
for act in activity:
    #両側検定，外れ値以外を出力する
    #print(act,":\n",grubbs.test(df[act], alpha=0.05),"\n")
    
    #片側検定,外れ値を出力する
    print(act,":\n",grubbs.min_test_outliers(df[act], alpha=0.05))
    print(act,":\n",grubbs.max_test_outliers(df[act], alpha=0.05))

```

```python
#どこかの群に有意差があるかということがわかる
#等分散の時に使える！
stats.f_oneway(df.eat, df.wipe, df.type, df.write, df.fold, df.peg)
```
