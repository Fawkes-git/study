```python
%matplotlib inline
import numpy as np
import pandas as pd
import re as re
import sklearn
import seaborn as sns
import matplotlib as plt
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os.path

import csv

df = pd.read_csv(os.path.join(os.getcwd(),'data/train21.tsv'), delim_whitespace=True)
print (df.info())
```

```python
data = pd.read_table(os.path.join(os.getcwd(),'data/distance.tsv'))
X = data[["Volt"]].values 
y = data["Dist"].values 
regr = LinearRegression()
regr = regr.fit(X, y)
y_fit = regr.predict(X)
print(regr.coef_)
print(regr.intercept_)
min = data["Dist"][0]
max = data["Dist"][1]
```

```python
volt = df["Volt"]
df["Dist"] =volt * regr.coef_ + regr.intercept_
df.plot(x="Time",y="Volt",figsize=(20,10),title='finger movement')
df.plot(x="Time",y="Dist",figsize=(20,10),title='finger movement')
```

```python
df["Dist"] = df["Dist"].apply(lambda x : min if x < min else (max if x > max else x))
df.plot(x="Time",y="Dist",figsize=(20,10),title='finger movement')
plt.xlabel("Time(s)")
plt.ylabel("Distance(mm)")
```

```python
df["Diff"] = df["Dist"].diff()
df["Fabs"]= np.fabs(df["Diff"])
df["Fabs_filter"] = df["Fabs"].apply(lambda x : 0 if x < 1 else  x)
df["Sum"] = df["Fabs_filter"].cumsum()
```

```python
df.plot(x="Time",figsize=(20,10),subplots=True);
plt.xlabel("Time(s)")
```
